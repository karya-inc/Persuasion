```python
import pandas as pd
import json

# Read CSV data using pandas
df = pd.read_csv('districts_states_mapped.csv')

# Group districts by state in Hindi
state_districts = df.groupby('pc11_state_name_hindi')['pc11_district_name_hindi'].apply(lambda x: x.str.strip().tolist()).to_dict()

# Convert the dictionary to JSON format
json_output = json.dumps(state_districts, ensure_ascii=False, indent=4)

# Optional: Save to a JSON file
with open('state_districts.json', 'w', encoding='utf-8') as f:
    f.write(json_output)
```

```python
# write all the unique states name sorted by name to a JSON file as JSON array
states = df['pc11_state_name_hindi'].unique().tolist()
states.sort()
json_output = json.dumps(states, ensure_ascii=False, indent=4)
with open('states.json', 'w', encoding='utf-8') as f:
    f.write(json_output)
    
```
